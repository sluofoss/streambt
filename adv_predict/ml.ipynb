```python
feature_target = db.sql(
f"""
select 
    TMF_w
    , TMF_4w_min
    , TMF_4w_min_dd
    , TMF_26w_min
    , TMF_26w_min_dd
    , TMF_4w_min_dd_qtl_50
    , TMF_4w_min_dd_qtl_50_alt
    , TMF_26w_min_dd_qtl_50
    , TMF_26w_min_dd_qtl_50_alt 
    
    , previous_exit_success
    , previous_exit_success2
    , previous_exit_result_ratio
    , previous_exit_result_ratio2
    --, TMF_Simple_Signal
    --, macd_hof
    --, macd_signal_hof
    --, macd_w_hof
    --, macd_w_signal_hof
    --, Volume
    , stddev_samp(log(Volume+1)) over (partition by Ticker order by Date rows between 5*52 preceding and current row) as std_vol
    , log(Volume + 1)
    --, dayofweek(Date) as day_of_week
    , Date
    , Ticker
    , gain_loss_ratio2
    , case 
        when gain_loss_ratio2 > 1.01 then 1 
        when gain_loss_ratio2 > 1.0 then 0
        else -1 
    end as target
from full_df
where TMF_Simple_Signal = 1
--and Ticker in string 
"""
)
train = db.sql("select * from feature_target where year(Date) in (2005,2007,2009,2011,2013,2015)").df() # worked ????

#train = db.sql("select * from feature_target where year(Date) in (2017,2018)").df()

print(train.columns)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score

clf = DecisionTreeClassifier(max_depth=5,class_weight={0:10,1:1}) # worked????

#clf = DecisionTreeClassifier(max_depth=5,class_weight={-1:20,0:10,1:1}) #
#clf = RandomForestClassifier(n_estimators= 10,max_depth=20)
#clf = RandomForestClassifier(n_estimators= 10,max_depth=50, class_weight={0:10,1:1})
#clf = RandomForestClassifier(n_estimators= 10,max_depth=10)
#clf = RandomForestClassifier(n_estimators= 10,max_depth=10, class_weight={0:10,1:1})
#clf = RandomForestClassifier(n_estimators= 2,max_depth=10)#, class_weight={0:10,1:1})
#clf = RandomForestClassifier(n_estimators= 50,max_depth=10, class_weight={0:10,1:1})

drop_columns = ['Date', 'Ticker', 'target', 'gain_loss_ratio2']
clf.fit(train.drop(columns = drop_columns),train['target'])

s_train = clf.score(train.drop(columns = drop_columns),train['target'])
print(s_train)
print(confusion_matrix(train['target'],clf.predict(train.drop(columns = drop_columns))))
print(precision_score(train['target'],clf.predict(train.drop(columns = drop_columns)), average = 'weighted' ))
```
